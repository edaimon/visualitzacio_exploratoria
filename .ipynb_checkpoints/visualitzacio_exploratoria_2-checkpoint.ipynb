```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('preu-lloguer-per trimestre-i-barri.csv')
df
```

```python
# Primera ullada a les dades
df.head(20)
```

No s'observen valors NaN. Comprovem-ho

```python
# Per contabilitzar els valors nuls a cada columna del dataset
print(df.isnull().sum())
```

No hi ha NaN!

```python
# Exploració dels tipus de dades del dataset
print(df.dtypes)
```

El preu hauria de ser un integer / float per fer càlculs amb la variable

```python
# Prova de si Seaborn agafa objectes per plotejar
sns.distplot(df['Preu'])
```

```python
# Canvi de datatype de 'Preu'
df['Preu'] = df['Preu'].astype(float)
```

Hi ha un valor '--' que no es pot convertir a float. 

```python
# Buscar valor '--' en 'Preu'
df.loc[df['Preu']=='--']
```

Per decidir què fer amb els valors '--', mirem el percentatge de casos sobre el total. Per això, primer s'han de transformar els valors a NaN

```python
# Transformació de '--' a NaN amb Numpy
df['Preu'] = df['Preu'].replace('--', np.nan)
```

```python
# Veiem el percentatge de NaNs
df['Preu'].isna().value_counts(normalize=True) * 100
```

Amb una proporció de 7.53 %, decideixo esborrar aquest registres del dataset


```python
# Back up del dataset
df2 = df
```

```python
# Esborrar files amb valors nuls
df2 = df.dropna()
```

```python
# Per contabilitzar els valors nuls a cada columna del dataset
df2.isnull().sum()
```

Nans ara son 0, procedim a convertir la columna a float

```python
# Canvi de datatype de 'Preu'
df2['Preu'] = df2['Preu'].astype(float)
```

```python
df2.dtypes
```

```python
df2
```

## Visualització de tres variables (Barri o Districte, Trimestre i Preu)

```python
df2['Any'].unique()
```

```python
# Crear la matriu de scatterplots

sns.relplot(x='Nom_Districte', y='Preu', col='Trimestre', hue='Nom_Districte', 
            data=df2_mitjana_preu_mensual)
plt.xticks(rotation=90)

```

```python
# Corregir els xticks per posar totes les etiquetes de l'eix x
scatter_preu_trimestre = sns.relplot(x='Nom_Districte', y='Preu', col='Trimestre', hue='Nom_Districte',
                                    data=df2)
plt.suptitle("Preu total de lloguer per trimestre i districte", fontsize = 26)
plt.subplots_adjust(top=0.8)
# Iterar per cada gràfica
for ax in scatter_preu_trimestre.axes.flat:
    ax.set_xticklabels(ax.get_xticklabels(), rotation=90)


# Desar gràfica en png

plt.savefig('scatter_preu_trimestre.png', dpi=300)

# Mostrar tots els gràfics
plt.show()

```

#### Amb els gràfics de dispersió per trimestre, veiem amb més detall la distribució de preus per districte, on també es reflexa la disparitat i pujada de preus màxims en Les Corts al T3 i Sant Martí al T4, per posar dos exemples evidents

## Proposa alguna visualització que creguis que pot resultar interessant.

Per fer visualtizacions més interssants, s'han de calcular algunes variables estadístiques a partir del dataset. Per exemple:

```python
# Convertir les dades a numèriques
df2_mitjana = df2_mitjana_preu_mensual
df2_mitjana['Preu'] = pd.to_numeric(df2_mitjana['Preu'])

# Calcular mitjana
df2_mean = df2_mitjana_preu_mensual.groupby(['Nom_Districte','Trimestre'])['Preu'].mean().round(2).reset_index()
df2_mean
```

```python
# Configuració de canvas
plt.figure(figsize=(8, 6))

sns.lineplot(data=df2_mean, x='Trimestre', y='Preu', hue='Nom_Districte')
plt.xlabel('Trimestre')
plt.ylabel('Preu en €')
plt.title('Evolució del preu per trimestre i districte (2021)')

# Correcció de posició de llegenda
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')

plt.show()
```

           Aquí es veu clarament la variació de preu per trimestre a cada districte

```python
# Configuració de canvas
plt.figure(figsize=(8, 6))

# Histograma acumulat (en stack), ens p
sns.histplot(data=df2_mitjana_preu_mensual, x="Preu", hue="Nom_Districte", multiple="stack")
plt.xlabel('Preu en €')
plt.ylabel('Registres')
plt.title('Distribució de registres per preu total de lloguer (2021)')

plt.show()
```

```python
# El Pairplot ens permeteix veure les dades des de diferents vistes, però aqui no afegeix informació útil
grafica_pairplot = df2_mitjana_preu_mensual[['Nom_Districte', 'Trimestre', 'Preu']]

sns.pairplot(grafica_pairplot, hue='Nom_Districte')
plt.savefig('grafica_pairplot.png', dpi=300)
```
