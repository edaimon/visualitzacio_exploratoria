```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('preu-lloguer-per trimestre-i-barri.csv')
df
```

```python
# Primera ullada a les dades
df.head(20)
```

No s'observen valors NaN. Comprovem-ho

```python
# Per contabilitzar els valors nuls a cada columna del dataset
print(df.isnull().sum())
```

No hi ha NaN!

```python
# Exploració dels tipus de dades del dataset
print(df.dtypes)
```

El preu hauria de ser un integer / float per fer càlculs amb la variable

```python
# Prova de si Seaborn agafa objectes per plotejar
sns.distplot(df['Preu'])
```

```python
# Canvi de datatype de 'Preu'
df['Preu'] = df['Preu'].astype(float)
```

Hi ha un valor '--' que no es pot convertir a float. 

```python
# Buscar valor '--' en 'Preu'
df.loc[df['Preu']=='--']
```

Per decidir què fer amb els valors '--', mirem el percentatge de casos sobre el total. Per això, primer s'han de transformar els valors a NaN

```python
# Transformació de '--' a NaN amb Numpy
df['Preu'] = df['Preu'].replace('--', np.nan)
```

```python
# Veiem el percentatge de NaNs
df['Preu'].isna().value_counts(normalize=True) * 100
```

Amb una proporció de 7.53 %, decideixo esborrar aquest registres del dataset


```python
# Back up del dataset
df2 = df
```

```python
# Esborrar files amb valors nuls
df2 = df.dropna()
```

```python
# Per contabilitzar els valors nuls a cada columna del dataset
df2.isnull().sum()
```

Nans ara son 0, procedim a convertir la columna a float

```python
# Canvi de datatype de 'Preu'
df2['Preu'] = df2['Preu'].astype(float)
```

```python
df2.dtypes
```

```python
df2
```

## Visualització d'una variable numèrica i una categòrica (Districte i Preu)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
preu = df2['Preu']
districte = df2['Nom_Districte']

sns.barplot(df2['Preu'], df2['Nom_Districte'])
plt.show()
```

```python
# Dibuixar gràfic
preu_districte = sns.barplot(x='Preu', y='Nom_Districte', data=df2)

# Configuracío de llegendas de la gràfica
plt.xlabel('Preu de lloguer')
plt.ylabel('Nom districte')

# Exportar png 
plt.savefig('preu-per-districte-mitj.png')

# Mostrar en pantalla
plt.show()

```

### Conclusió: en la gràfica es comprova que els districtes amb preus més cars són Les Corts i Sarrià-Sant Gervasi, i els més barats Sant Andreu i Nou Barris.
#### Per millorar la lectura i analitzar les mitjanes, ordenarem les barres de preu més alt a més baix

```python
# Ordenem de preu més alt a més baix
df2_ordre = df2.sort_values(by='Preu', ascending=False)

# Dibuixar gràfic
preu_districte = sns.barplot(x='Preu', y='Nom_Districte', data=df2_ordre)

# Configuració de llegendas de la gràfica
plt.xlabel('Preu de lloguer')
plt.ylabel('Nom districte')

# Exportar png 
plt.savefig('preu-per-districte-mitj-ord.png')

# Mostrar en pantalla
plt.show()
```

### Conclusió: <br>Aquí es veu millor que mentres Les Corts té els preus més alts, la mitjana de Sarrià-Sant Gervasi és més alta, és a dir, els preus varien menys a aquest barri<br>


## Visualització de dues variables numèriques (Any o Trimestre i Preu)


```python
# Prova amb scatterplot

trim_preu = sns.scatterplot(data=df2, x='Trimestre', y='Preu', color = "magenta")
plt.show()
```

### En aquesta figura, realitzada amb un gràfic de dispersió (scatter plot), les dades estan massa alineades i no es pot apreciar adequadament el volum de registres.

```python
# Prova amb stripplot

trim_preu = sns.stripplot(data=df2, x='Trimestre', y='Preu', color = "magenta", alpha=0.7)
plt.savefig('trim_preu_stripplot.png')
plt.show()
```

### Amb l'stripplot, que dibuixa cada registre, es veu millor l'acumuació de registres per trimestre

## Visualització de tres variables (Barri o Districte, Trimestre i Preu)

```python
df2['Any'].unique()
```

```python
# Crear la matriu de scatterplots

sns.relplot(x='Nom_Districte', y='Preu', col='Trimestre', hue='Nom_Districte', 
            data=df2)
plt.xticks(rotation=90)

```

```python
# Corregir els xticks per posar totes les etiquetes de l'eix x
scatter_preu_trimestre = sns.relplot(x='Nom_Districte', y='Preu', col='Trimestre', hue='Nom_Districte',
                                    data=df2)

# Iterar per cada gràfica
for i in scatter_preu_trimestre.axes.flat:
    i.set_xticklabels(i.get_xticklabels(), rotation=90)

# Desar gràfica en png
plt.savefig('scatter_preu_trimestre.png')

# Mostrar tots els gràfics
plt.show()

```
